## Projectile data

```python
import numpy as np
from numpy import random
import matplotlib.pyplot as plt
```

```python
#np.random.seed(15)
npts = 100
x = np.random.randn(npts)*1.5
x = x - np.min(x)+.05
eps = np.random.randn(npts)*50
vel = 300
ht = 500

y = -0.5*9.8*x**2 + vel*x + ht + eps
plt.plot(x,y,'.')
```

```python
import pandas as pd
```

```python
df = pd.DataFrame({'x':x, 'y':y})
```

```python
df.to_csv('data.csv', index=False)
```

```python
import seaborn as sns
```

```python
sns.regplot(x,y, line_kws={'color': 'red'})
```
